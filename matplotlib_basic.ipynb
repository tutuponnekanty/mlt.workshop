## _Plotting Graphs using Numpy and Matplotlib_

```python
#importing our required functions

import numpy as np

import matplotlib.pyplot as plt #we don't need the full library here
```

```python
#representing a real line in vector format using linspace
new_line=np.linspace(0,1,3) #(startpoint, endpoint, total partitions inclusive of both start and end)
print(new_line)
```

```python
x=np.linspace(1,100,5,endpoint=False) #just neglecting the endpoint, but the values will be generated considering the end point as the limit
#note : we cannot neglect the start point in numpy
```

```python
y=x+1
y
```

```python
#plotting using Matplotlib
plt.plot(x,y)
```

```python
plt.scatter(x, y, color='red')
plt.plot(x,y)
#visualizing the points using scatter plot
```

```python
def random_plot():
    x=np.linspace(-5,5,20)
    y=5-3*x
    return(x,y)
x, y = random_plot()
plt.plot(x, y)
plt.scatter(x,y)
plt.title('graph representation')
```

```python
x=np.linspace(-1,3,100)
y=x**2
plt.plot(x,y)
```

```python
#Coloring using matplotlib
#we useplot(x, y, color='green', marker='o', linestyle='dashed',... linewidth=2, markersize=12)
```

```python
# y = 3
x = np.linspace(0, 6, 2)
y = np.ones(2) * 3
plt.plot(x, y, color = 'blue', label = 'y = 3')

# x = 5
x = np.ones(2) * 5
y = np.linspace(0, 6, 2)
plt.plot(x, y, color = 'red', label = 'x = 5')

plt.legend()
plt.title('x = 5 and y = 3')
plt.axhline(color = 'black')
plt.axvline(color = 'black');
```
